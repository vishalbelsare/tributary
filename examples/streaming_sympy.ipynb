```python
import tributary.symbolic as ts
import tributary.streaming as tss
```

```python
expr = ts.parse_expression("10sin**2 x**2 + 3xyz + tan theta")
```

```python
clz = ts.construct_streaming(expr)
```

```python
def func(*args):
    for _ in range(5):
        yield _

timer = tss.Print(tss.Timer(lambda: 1, count=0))
x = clz(x=tss.Const(1), y=tss.Func(func), z=timer, theta=tss.Const(4))
```

```python
y = tss.run(tss.Print(x._node))
y
```
